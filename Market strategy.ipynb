This dataset taken from ***Marketing_Campaign_Data_A.csv***

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('Marketing_Campaign_Data_A.csv')
df.head(10)
```

```python
df = df.iloc[: , 1:]
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['DATE_BIRTH'] = pd.to_datetime(df.DATE_BIRTH)
df['APPL_DATETIME'] = pd.to_datetime(df.APPL_DATETIME)
df
```

```python
df['Age'] = (df.APPL_DATETIME - df.DATE_BIRTH).astype('timedelta64[Y]').astype('int')
df
```

```python
df.loc[df['Age']<20, 'age_group'] = 'Late_Ten'
df.loc[df['Age'].between(20,30), 'age_group'] = 'Twenty'
df.loc[df['Age'].between(30,40), 'age_group'] = 'Thirty'
df.loc[df['Age'].between(40,50), 'age_group'] = 'Fourty'
df.loc[df['Age'].between(50,60), 'age_group'] = 'Fifty'
df.loc[df['Age']>=60, 'age_group'] = 'Above_Sixty'
```
